```python
import math

i = 0
k = 2500 
samples = []
filtered_lan = 0
filtered_gopher = 0
with open("data/CC/example.warc.gz", "rb") as f:
    iterator = tqdm(ArchiveIterator(f))
    for record in iterator:
        binary_data = record.reader.read()
        text = run_extract_text_from_html_bytes(binary_data)
        lang_res = run_identify_language(text)

        iterator.set_postfix(kept=len(samples), lang=filtered_lan, gopher=filtered_gopher)
        
        if lang_res[0] != 'en' or lang_res[1] < 0.6:
            filtered_lan+=1
            continue

        if not run_gopher_quality_filter(text):
            filtered_gopher+=1
            continue
            
        i+=1
        
        if i <=k:
            samples.append(text)
        else:
            j = math.floor(random.random() * i)
            if j < k:
                samples[j] = text
                
```

```python
filtered_lan = 0
filtered_gopher = 0
positives = []
with open("data/wiki/positives.warc.gz", "rb") as f:
    iterator = ArchiveIterator(f)
    for record in iterator:
        ct = record.http_headers.get("Content-Type", "") if record.http_headers else ""
        if not ct.startswith("text/html"):
            continue
        
        binary_data = record.reader.read()
        text = run_extract_text_from_html_bytes(binary_data)

        lang_res = run_identify_language(text)

        if lang_res[0] != 'en' or lang_res[1] < 0.6:
            filtered_lan+=1
            continue

        if not run_gopher_quality_filter(text):
            filtered_gopher+=1
            continue

        positives.append(text)
    
```
